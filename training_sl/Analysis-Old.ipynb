```python
def read_track(files):
    data = []
    for track_file in files:
         
        with open(track_file, "rb") as f:
            d = pickle.load(f)
        data.extend(d)
    return data

def read_model(files):
    files_to_read = []
    data = []
    _ = [files_to_read.extend(f) for f in files]
    for f in files_to_read:
        with open(f, "rb") as f:
            d = pickle.load(f)
        data.extend(d)
    return data
```
