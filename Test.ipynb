```python
import numpy as np
from ddpg import Trainer
```

```python
states = np.load("state.npy")
next_states = np.load("state.npy")
rewards = np.load("reward.npy")
actions = np.load("actions.npy")
```

```python
trainer = Trainer()
```

```python
trainer.buffer.state_buffer = states
trainer.buffer.action_buffer = actions
trainer.buffer.reward_buffer = rewards
trainer.buffer.next_state_buffer = next_states
```

```python
trainer.buffer.buffer_counter = 500
```

```python
for i in range(10000):
    loss, grad = trainer.buffer.learn(trainer.actor_model, trainer.critic_model,
                        trainer.target_actor, trainer.target_critic, trainer.gamma, trainer.critic_optimizer,
                                     trainer.actor_optimizer)
    print("Loss: ", loss.numpy())
#     print([np.median(l) for l in grad])
```
